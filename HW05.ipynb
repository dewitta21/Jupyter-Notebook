Question 2
Write a while loop that prints all positive numbers that are divisible by 10 and less than a given number n. For example, if n is 100, print 10 20 30 40 50 60 70 80 90, one value per line.

```python
n = int(input("n: "))
start = 1
while start < n :
   if start % 10 == 0:
      print(start)
   start = start + 1
```
