# question 1
It is March 14, and the bakery is holding a pie sale. All pies with a regular price of $10 or more are $3.14 off. All others are $1.00 off. Write a program that reads the price of a pie and prints the discounted price.

```python
#question 1
userInput = input("Original price: ")
price = float(userInput)

if price >= 10.00 :
   discountedPrice = price - 3.14
else :
   discountedPrice = price - 1.00        

print("Discounted price: %.2f\n" % discountedPrice)
```

# question 5
You are given the width and height of a picture. Determine whether it is in landscape orientation (wider than tall), in portrait orientation (taller than wide), or square, then print Landscape, Portrait, or Square.

```python
width = int(input("Width: "))
height = int(input("Height: "))

if width > height :
   print("Landscape")
elif width == height :
   print("Square")
else :
   print("Portrait")
```
